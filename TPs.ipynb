#### Calculer la droite de régression de la distance par rapport é la vitesse par la méthode des moindres carrés



```r
df<- data.frame(vitesses, distances)
droite1<- lm(distances ~ vitesses, df)
coef(droite1)
```

#### Calculer la droite de régression de la distance par rapport é la distance



```r
droite2 <- lm(vitesses ~ distances, df)
coef(droite2)
```

```r
plot(vitesses ~ distances, df)
abline(droite2)
```

#### Donner une représentation graphique des deux droites et calculer l'angle qu'elles font entre ellles.



```r
plot(vitesses ~ distances, df)
abline(droite1)
abline(droite2)
```

```r
y1<- coef(droite1)
y2<- coef(droite2)
b1<- y1[2]
a1<- y1[1]
b2<- y2[2]
a2<- y2[1]
angle<- abs((a2-a1)/(1+a1*a2))
angle_rad=atan(angle)
angle_rad
angle_deg= angle_rad*(180/pi)
angle_deg
```

```r
cor(df$vitesses, df$distances)
```

```r
mydata = data.frame(x1 = c( 0, 0, 1, 1 ),
                    x2 = c( 0, 1, 0, 1 ),
                    y =  c( 0, 1, 1, 1 ) )

print(mydata)
```

```r
weights = rnorm(mean = 0, sd = 0.1, n = ncol(mydata))

print( weights )
```

```r
activ_function = function(net){
       if(net>0.5)
       return(1)
       return(0)
       }
```

```r
data = as.matrix(mydata)
net = c(data[1, 1:2 ], 1)%*% weights

y_hat = activ_function( net )

error = y_hat - data[1,3]
# eta est le taux d'apprentissage, cette valeur définit le pas pour parcourir la fonction d'erreur.
eta = 0.1

weights = weights-eta*( error )*c(data[ 1, 1:2 ], 1)

print(weights)
```

```r
#Cette fonction trouve les poids optimaux pour le modéle perceptron.
nn = function(dataset, eta = 0.1, threshold = 1e-5){
                        data = as.matrix(dataset)
                        num.features = ncol(data) - 1
                        target = ncol(data)

                        # Initial random  weights
                        weights = rnorm(mean = 0, sd = 0.1, n = ncol(data))

                        mse = threshold * 2
                        while( mse > threshold ){
                            mse = 0
                            for(i in 1:nrow(data)){
                                # Add bias and compute multiplications
                                net = c(data[ i, 1:num.features], 1) %*% weights

                                # Activation function
                                y_hat = activ_function(net)

                                # Compute mse
                                error = (y_hat - data[ i, target])
                                mse = mse + error^2
                                cat(paste("Mean square error = ", mse, "\n"))

                                # Update weights
                                weights = weights- eta*error*c(data[i, 1:num.features], 1)
                            }
                        }
                        return(weights)
                    }
```

```r

shattering.plane = function(weights){
                        X = seq(0, 1, length = 100 )
                        data = outer(X, X, function( X, Y ){cbind(X, Y, 1) %*% weights})
                        id = which(data>0.5)
                        data[id] = 1
                        data[-id] = 0
                        filled.contour(data)
                        }

weights = nn(data, eta = 0.1, threshold = 1e-5 )

shattering.plane(weights)
```
